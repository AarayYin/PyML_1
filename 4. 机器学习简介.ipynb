```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20, 5)
y = 2 * x + 1
```

```python
x
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y, s = 200, c = 'green')
```

```python
plt.xlim((0,30))
plt.ylim((0,50))
plt.scatter(x, y, s = 200, c = 'green', marker = ',')
```

```python
plt.xlim((0,30))
plt.ylim((0,50))

x_ticks = np.linspace(0, 30, 11)
plt.xticks(x_ticks)

plt.scatter(x, y, s = 200, c = 'green', marker = ',')
```

```python
plt.xlim((0,30))
plt.ylim((0,50))

x_ticks = np.linspace(0, 30, 11)
plt.xticks(x_ticks)

plt.xlabel('x0')
plt.ylabel('x0')

plt.scatter(x, y, s = 200, c = 'green', marker = ',')
```

```python
x = np.linspace(-10, 10, 40)
y = x * x + 1
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 40)
y = x * x + 1
y0 = -x * x

plt.plot(x, y)
plt.plot(x, y0, c = 'red')
```

```python
x = np.linspace(-10, 10, 40)
y = x * x + 1

x2 = np.linspace(-10, 10, 5)
y2 = 3 * x2 + 50

plt.scatter(x2, y2, c = 'red')
plt.plot(x, y)
```

```python
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)
Z = (1 - X/2 + X**3 + Y**4)*np.exp(-X**2-Y**2)
plt.pcolormesh(X, Y, Z)
```

```python
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)
Z = (1 - X/2 + X**3 + Y**4)*np.exp(-X**2-Y**2)
plt.contourf(X, Y, Z)
```
