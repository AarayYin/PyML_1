```python
from sklearn import neighbors
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
```

```python
with open('knn.csv', encoding = 'UTF-8') as f:
    names = ['x','y', 'labels']
    data = read_csv(f, names = names)
data
```

```python
x1 = data['x'].iloc[:200,]
y1 = data['y'].iloc[:200,]
x2 = data['x'].iloc[201:400,]
y2 = data['y'].iloc[201:400,]
x3 = data['x'].iloc[401:600,]
y3 = data['y'].iloc[401:600,]
plt.scatter(x1, y1, c = 'b', marker = 's', s = 50)
plt.scatter(x2, y2, c = 'r', marker = '^', s = 50)
plt.scatter(x3, y3, c = 'g', s = 50)
```

```python
excluded = ['labels']
X = data.drop(excluded, axis = 1)
y = data['labels'].values
print(X)
print(y)
```

```python
clf = neighbors.KNeighborsClassifier(30)
clf.fit(X,y)
```

```python
xx, yy = np.meshgrid(np.arange(0.2, 1.31, 0.01), np.arange(0.2, 1.31, 0.01))
```

```python
coords = np.c_[xx.ravel(), yy.ravel()]
```

```python
x0, y0 = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 10, 1))
print(x0)
print(x0.ravel())
print(y0.ravel())
```

```python
x_temp = np.array([[1,2,3],[4,5,6]])
y_temp = np.array([[7,8,9],[10,11,12]])
print(x_temp)
print(y_temp)
print(np.c_[x_temp,y_temp])
print(np.r_[x_temp,y_temp])
```

```python
print(coords)
```

```python
Z = clf.predict(coords)
```

```python
print(xx.shape)
```

```python
print(Z.shape)
```

```python
Z = Z.reshape(xx.shape)
print(Z.shape)
```

```python
plt.pcolormesh(xx, yy, Z)
plt.scatter(x1, y1, c = 'b',marker = 's', s = 50, alpha = 0.8)
plt.scatter(x2, y2, c = 'r', marker = '^', s = 50, alpha = 0.8)
plt.scatter(x3, y3, c = 'g', s = 50, alpha = 0.8)
plt.axis((0.2, 1.3, 0.2, 1.3))
```

```python
from matplotlib.colors import ListedColormap

light_rgb = ListedColormap(['#AAAAFF','#FFAAAA','#AAFFAA'])

plt.pcolormesh(xx, yy, Z, cmap = light_rgb)
plt.scatter(x1, y1, c = 'b',marker = 's', s = 50, alpha = 0.8)
plt.scatter(x2, y2, c = 'r', marker = '^', s = 50, alpha = 0.8)
plt.scatter(x3, y3, c = 'g', s = 50, alpha = 0.8)
plt.axis((0.2, 1.3, 0.2, 1.3))
plt.rcParams['figure.figsize']=(80, 80)
plt.savefig('test.jpg')
```
